# Day 1: 사인파(Sine Wave) 기초

오늘 목표:

- 진폭(amplitude), 주파수(frequency) 구분하기
- 시간축에서 100Hz/300Hz 파형 차이 보기
- 같은 주파수에서 진폭만 바꿔 비교하기

## 핵심 개념

사인파는 시간에 따라 부드럽게 반복되는 기본 파형입니다.

- 수식: `x(t) = A * sin(2πft)`
- `A` (진폭): 파형의 높이(세기)
- `f` (주파수): 1초에 반복되는 횟수(Hz)

시간축 그래프에서 주파수가 높을수록 같은 시간에 더 많은 진동이 보입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# 기본 설정
fs = 5000  # 샘플링 주파수(Hz)
duration = 0.03  # 30ms 구간만 관찰
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

f1, f2 = 100, 300
amp = 1.0

wave_100 = amp * np.sin(2 * np.pi * f1 * t)
wave_300 = amp * np.sin(2 * np.pi * f2 * t)
```

```python
# 100Hz vs 300Hz 비교 (같은 시간축)
plt.figure(figsize=(10, 4))
plt.plot(t, wave_100, label='100Hz, A=1.0', linewidth=2)
plt.plot(t, wave_300, label='300Hz, A=1.0', linewidth=2, alpha=0.5)
plt.title('100Hz와 300Hz 사인파 비교')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 같은 주파수(100Hz)에서 진폭만 바꾸기
base_freq = 100
amplitudes = [0.2, 0.5, 1.0]

plt.figure(figsize=(10, 4))
for a in amplitudes:
    wave = a * np.sin(2 * np.pi * base_freq * t)
    plt.plot(t, wave, label=f'100Hz, A={a}', linewidth=2)

plt.title('같은 주파수(100Hz)에서 진폭 변화')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.tight_layout()
plt.show()
```
